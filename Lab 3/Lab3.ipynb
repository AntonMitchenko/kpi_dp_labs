# 10. Розпізнавання мережевої кібератаки типу phf на базі PNN.

```python
df['label'].value_counts()
```

```python
data = df
data
```

```python
label_encoder = LabelEncoder()
```

```python
for col in ['service', 'flag', 'protocol_type']:
    data[col] = label_encoder.fit_transform(data[col])
```

```python
data
```

```python
X = data.drop('label', axis=1)
y = data['label']

class_to_perceive = 'phf'
y_binary = (y == class_to_perceive).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
class PNN:
    def __init__(self, std):
        self.std = std
        self.class_means = None
        self.class_probs = None

    def gauss_kernel(self, x, mean):
        return np.exp(-0.5 * np.sum(((x - mean) / self.std) ** 2))

    def train(self, X, y):
        self.class_means = [np.mean(X[y == 1], axis=0), np.mean(X[y == 0], axis=0)]
        self.class_probs = [len(X[y == 1]) / len(X), len(X[y == 0]) / len(X)]

    def predict_instance(self, x):
        class_scores = []

        for mean, prob in zip(self.class_means, self.class_probs):
            class_score = prob * self.gauss_kernel(x, mean)
            class_scores.append(class_score)

        return np.argmax(class_scores)

    def predict(self, X):
        predictions = [self.predict_instance(x) for x in X]
        return np.array(predictions)
```

```python
start_time = time.time()

pnn = PNN(std=0.1)
pnn.train(X_train_scaled, y_train)

end_time = time.time()

predictions = pnn.predict(X_test_scaled)
```

```python
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy:.2f}")

print("Classification Report:")
print(classification_report(y_test, predictions))

training_duration = end_time - start_time
print(f"Training Duration: {training_duration:} seconds")
```
